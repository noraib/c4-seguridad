# MNIST

```python
from datasets import load_dataset
import numpy as np

ds = load_dataset("ylecun/mnist")
```

```python
# 2️⃣ Chequear cantidad de ejemplos en train/test
print(f"Train samples: {len(ds['train'])}, Test samples: {len(ds['test'])}")

# 3️⃣ Tomar una imagen de ejemplo
img = ds['train'][0]['image']  # PIL Image
print(f"Tipo de objeto: {type(img)}")

# 4️⃣ Convertir a numpy array para analizar píxeles y canales
img_np = np.array(img)
print(f"Shape de la imagen: {img_np.shape}")
print(f"Valores min/max: {img_np.min()}/{img_np.max()}")
print(f"Cantidad de canales: {1 if len(img_np.shape) == 2 else img_np.shape[2]}")
print(f"Total de píxeles: {img_np.shape[0] * img_np.shape[1]}")

# 5️⃣ Revisar estadísticas básicas del dataset
all_shapes = [np.array(x['image']).shape for x in ds['train']]
unique_shapes = set(all_shapes)
print(f"Formas únicas de imágenes en el dataset: {unique_shapes}")
```
